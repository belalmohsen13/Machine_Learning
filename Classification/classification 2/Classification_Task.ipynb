### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).

# Data Cleaning & Preprocessing

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
y=y.map({'p':1,'e':0})
```

```python
X_cat=X.select_dtypes(include='object').columns
X_num=X.select_dtypes(include=['float64', 'int64']).columns
```

```python
X_encoded=pd.get_dummies(columns=X_cat,
                          data=X,
                          drop_first=True)
```

```python
X_encoded.head()
```

```python
scaler=StandardScaler()
X[X_num]=scaler.fit_transform(X_encoded[X_num])
```

```python
X_encoded.head(5)
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.3,random_state=42,stratify=y)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

print("After SMOTE:", y_train.value_counts())
```

# feature selection

```python

from sklearn.feature_selection import SelectKBest, chi2, f_classif, mutual_info_classif

# Chi-Square test (categorical ↔ target)
chi2_selector = SelectKBest(score_func=chi2, k=10)  # top 10 features
chi2_selector.fit(X_train, y_train)
chi2_features = X_train.columns[chi2_selector.get_support()]
print("Top 10 features by Chi-Square Test:")
print(chi2_features.tolist())

# ANOVA F-test (numerical ↔ target)
anova_selector = SelectKBest(score_func=f_classif, k=10)
anova_selector.fit(X_train, y_train)
anova_features = X_train.columns[anova_selector.get_support()]
print("\nTop 10 features by ANOVA F-test:")
print(anova_features.tolist())

# Mutual Information
mi_selector = SelectKBest(score_func=mutual_info_classif, k=10)
mi_selector.fit(X_train, y_train)
mi_features = X_train.columns[mi_selector.get_support()]
print("\nTop 10 features by Mutual Information:")
print(mi_features.tolist())

# Combine results in a DataFrame for comparison
scores = pd.DataFrame({
    "Feature": X_train.columns,
    "Chi2": chi2_selector.scores_,
    "ANOVA": anova_selector.scores_,
    "MI": mi_selector.scores_,
})

print("\n=== Feature Ranking (Top Scores) ===")
print(scores.sort_values("Chi2", ascending=False).head(15))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
model = LogisticRegression(max_iter=100)
rfe=RFE(model,n_features_to_select=10)
rfe.fit(X_train,y_train)
rfe.get_feature_names_out()
rfe_features = set(X_train.columns[rfe.get_support()])
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
rf=RandomForestClassifier(random_state=42)
rf.fit(X_train,y_train)
importances = rf.feature_importances_
rf_features = X_train.columns[np.argsort(importances)[-20:]]
```

```python
all_features = set().union(chi2_features, anova_features, mi_features, rfe_features, rf_features)

comparison = pd.DataFrame(index=sorted(all_features))
comparison["Chi2"] = comparison.index.isin(chi2_features).astype(int)
comparison["ANOVA"] = comparison.index.isin(anova_features).astype(int)
comparison["Mutual_Info"] = comparison.index.isin(mi_features).astype(int)
comparison["RFE"] = comparison.index.isin(rfe_features).astype(int)
comparison["RandomForest"] = comparison.index.isin(rf_features).astype(int)

# Add a column: total times feature was selected
comparison["Total"] = comparison.sum(axis=1)

# Sort by most frequently selected
comparison = comparison.sort_values("Total", ascending=False)

print("\n=== Feature Selection Comparison ===")
print(comparison)
```

```python
top10_features = comparison.sort_values("Total", ascending=False).head(10).index.tolist()

print("Top 10 Selected Features:")
print(top10_features)
```

```python
X_train_top10 = X_train[top10_features]
X_test_top10  = X_test[top10_features]
```

# 5. Model Training & Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
import seaborn as sns
import matplotlib.pyplot as plt


models = {
    "Logistic Regression": LogisticRegression(max_iter=500, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42, probability=True),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

results = {}

for name, model in models.items():
    print(f"\n===== {name} =====")
    
    model.fit(X_train_top10, y_train)
    y_pred = model.predict(X_test_top10)
    
   
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy:", acc)
    print("Classification Report:\n", classification_report(y_test, y_pred))
    
   
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    cv_scores = cross_val_score(model, X_train_top10, y_train, cv=5)
    print("Cross-validation mean accuracy:", cv_scores.mean())
    
    results[name] = {
        "Accuracy": acc,
        "CV Mean": cv_scores.mean()
    }

# جدول مقارنة النتايج
results_df = pd.DataFrame(results).T
print("\n=== Model Comparison ===")
print(results_df)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# 1) Decision Tree
dt_params = {
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 5, 10]
}
dt_grid = GridSearchCV(DecisionTreeClassifier(random_state=42),
                       dt_params, cv=5, scoring="accuracy")
dt_grid.fit(X_train_top10, y_train)

print("=== Decision Tree ===")
print("Best Params:", dt_grid.best_params_)
print("Best CV Score:", dt_grid.best_score_)

y_pred_dt = dt_grid.best_estimator_.predict(X_test_top10)
print("Test Accuracy:", accuracy_score(y_test, y_pred_dt))


# 2) Random Forest
rf_params = {
    "n_estimators": [50, 100],
    "max_depth": [None, 10],
}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42),
                       rf_params, cv=5, scoring="accuracy")
rf_grid.fit(X_train_top10, y_train)

print("\n=== Random Forest ===")
print("Best Params:", rf_grid.best_params_)
print("Best CV Score:", rf_grid.best_score_)

y_pred_rf = rf_grid.best_estimator_.predict(X_test_top10)
print("Test Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(10, 8))

models = {
    "Logistic Regression": LogisticRegression(max_iter=500, class_weight="balanced"),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(class_weight="balanced", random_state=42),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True, random_state=42)  # Added SVM properly
}

for name, model in models.items():
    model.fit(X_train[top10_features], y_train)
    
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test[top10_features])[:, 1]
    else:
        y_proba = model.decision_function(X_test[top10_features])
    
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Diagonal line
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance (AUC = 0.50)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
