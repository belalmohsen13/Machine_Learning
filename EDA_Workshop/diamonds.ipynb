# vis cut and Price and color

```python
sns.barplot(x='cut', y='price', hue= 'color',data=df)
plt.title('Cut vs Price')
plt.show()
```

```python
df.head(5)
df['cut'] = df['cut'].replace({'GoodVery': 'Very Good','ideal' : 'Ideal','premium':'Premium'})
df = df[df['cut'] != 'Others']
```

```python
sns.barplot(x='cut', y='price', hue= 'color',data=df)
plt.title('Cut vs Price')
plt.show()
```

Better cut quality (Ideal, Premium) leads to higher diamond prices. Lower cut grades (Fair, Good) result in lower prices.

# Scaling

```python
num=df.select_dtypes(include=[int,float])
num
```

## first scaling

```python
df['carat'].plot(kind='hist')
```

```python
from sklearn.preprocessing import PowerTransformer
pw=PowerTransformer(method='box-cox')
df['carat_scaled']=pw.fit_transform(df[['carat']])
```

```python
plt.subplot(1,2,1)
df['carat'].plot(kind='hist')
plt.subplot(1,2,2)
df['carat_scaled'].plot(kind='hist')

plt.tight_layout()
```

## second scaling

```python
df['depth'].plot(kind='hist')
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df['depth_scaled']=ss.fit_transform(df[['depth']])
```

```python
plt.subplot(1,2,1)
df['depth'].plot(kind='hist')
plt.subplot(1,2,2)
df['depth_scaled'].plot(kind='hist')

plt.tight_layout()
```

## third scaling

```python
df['table'].plot(kind='hist')
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df['table_scaled']=ss.fit_transform(df[['table']])
```

```python
plt.subplot(1,2,1)
df['table'].plot(kind='hist')
plt.subplot(1,2,2)
df['table_scaled'].plot(kind='hist')

plt.tight_layout()
```

## fourth scaling

```python
df['price'].plot(kind='hist')
```

```python

df['price_scaled']=pw.fit_transform(df[['price']])
```

```python
plt.subplot(1,2,1)
df['price'].plot(kind='hist')
plt.subplot(1,2,2)
df['price_scaled'].plot(kind='hist')

plt.tight_layout()
```

## fifth scaling

```python
df['z'].plot(kind='hist')
df['y'].plot(kind='hist',alpha=0.3)

```

```python
pwy = PowerTransformer(method='yeo-johnson')
df['z_scaled'] = pwy.fit_transform(df[['z']])
df['y_scaled'] = pwy.fit_transform(df[['y']])

df['z_scaled'].plot(kind='hist', alpha=0.5, label='z_scaled')
df['y_scaled'].plot(kind='hist', alpha=0.5, label='y_scaled')
plt.legend()
```
