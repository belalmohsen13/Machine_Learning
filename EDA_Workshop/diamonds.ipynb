# vis Carat and Price'

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['price'], kde=True)
plt.title('Price')
plt.show()
```

```python
sns.boxplot(df['carat'])
plt.title('Carat')
plt.show()
```

```python
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Carat vs Price')
plt.show()
```

```python
sns.lmplot(x='carat', y='price', data=df)
plt.title('Carat vs Price')
plt.show()
```

```python
sns.regplot(x='carat', y='price', data=df)
plt.title('Carat vs Price')
plt.show()
```

```python
sns.lineplot(x='carat', y='price', data=df)
plt.title('Carat and Price')
plt.show()
```
